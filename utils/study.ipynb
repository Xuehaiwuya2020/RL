```python
import torch.nn as nn
import torch
import torch.nn.functional as F
import torch.optim as optim

class SimpleNet(nn.Module):
    def __init__(self, input_dim = 16, hidden_size=64):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_size)
        self.ac1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.ac1(x)
        x = self.fc2(x)
        return x

x = torch.randn(8, 16)
y = torch.randn(8, 1)


model = SimpleNet()
optimizer = optim.Adam(model.parameters(), lr=1e-2)
for epoch in range(50):
    print(f"Epoch {epoch}")
    pred = model(x)

    loss_func = nn.MSELoss()
    loss = loss_func(pred, y)


    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    

    print(loss.item())

print("训练完成")
```

```python
class miniPolicyNet(nn.Module):
    def __init__(self, in_dim = 4, hidden_dim = 8):
        super().__init__()
        self.fc1 = nn.Linear(in_dim, hidden_dim)
        self.act1 = nn.ReLU()
        self.ac = nn.Linear(hidden_dim, 2)
        self.act2 = nn.Softmax(dim=-1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.ac(x)
        x = self.act2(x)
        return x
    

model = miniPolicyNet()
optimizer = optim.Adam(model.parameters(), lr=1e-2)  # new environment
state = torch.randn(1, 4)

for epoch in range(100):
    print(f"epoch {epoch}")
    pred = model(state) # size [batch_size, 2]
    
    dist = torch.distributions.Categorical(pred)
    action = dist.sample()
    # print(type(action), action.item())
    log_prob = dist.log_prob(action)
    if action.item() == 0:
        reward = 1.0
    else:
        reward = -1.0
    optimizer.zero_grad()
    loss = -log_prob * reward
    loss.backward()
    optimizer.step()

    print(loss.item())

print("training complete")
```

```python
class miniPolicyNet(nn.Module):
    def __init__(self, in_dim = 4, hidden_dim = 8):
        super().__init__()
        self.fc1 = nn.Linear(in_dim, hidden_dim)
        self.act1 = nn.ReLU()
        self.ac = nn.Linear(hidden_dim, 2)
        self.act2 = nn.Softmax(dim=-1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.ac(x)
        x = self.act2(x)
        return x
    
class miniValueNet(nn.Module):
    def __init__(self, in_dim = 4, hidden_dim = 8):
        super().__init__()
        self.fc1 = nn.Linear(in_dim, hidden_dim)
        self.act1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, 1)
    
    def forward(self, x):
        return self.fc2(self.act1(self.fc1(x)))

model = miniPolicyNet()
optimizer = optim.Adam(model.parameters(), lr=1e-2)  # new environment

v_model = miniValueNet()
v_optimizer = optim.Adam(v_model.parameters(), lr=1e-2)

v_loss_func = nn.MSELoss()  # value loss function

for epoch in range(200):
    print(f"epoch {epoch}")
    state = torch.randn(1, 4)
    # state action
    pred = model(state) ## size [batch_size, 2] 
    # print(state[0][0].item(), pred.detach())
    dist = torch.distributions.Categorical(pred)
    action = dist.sample()
    v_state = v_model(state)
    # print(type(action), action.item())
    log_prob = dist.log_prob(action)
    if state[0][0] > 0:
        if action.item() == 0:
            reward = 1.0
        else:
            reward = -1.0
    else:
        if action.item() == 0: # 
            reward = -1.0
        else:
            reward = 1.0
    
    optimizer.zero_grad()
    v_optimizer.zero_grad()
    loss = -log_prob * (reward - v_state.detach())
    v_loss = v_loss_func(v_state, torch.tensor([[reward]]))

    loss = loss + v_loss
    loss.backward()
    optimizer.step()
    v_optimizer.step()
    
    print(loss.item())

print("training complete")
```

```python
"""
转换数据格式
"""
import json


def convert_dialogue(input_file, output_file):
    """
    转换对话格式
    
    输入：带attrs的原始对话数据（JSON数组）
    输出：Qwen格式的多轮对话（JSONL，每行一条）
    """
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    with open(output_file, 'w', encoding='utf-8') as fout:
        for dialogue in data:
            messages = dialogue.get("messages", [])
            
            conversations = []
            
            # 遍历每条消息
            cnt = 0
            for msg in messages:
                message_text = msg.get("message", "").strip()
                
                if not message_text:
                    continue
                
                # 判断role：有attrs的是assistant，没有的是user
                if cnt % 2 == 1:
                    role = "assistant"
                else:
                    role = "user"
                
                conversations.append({
                    "role": role,
                    "content": message_text
                })
                cnt += 1
            output_sample = {
                "conversations": conversations
            }
            fout.write(json.dumps(output_sample, ensure_ascii=False) + '\n')


if __name__ == "__main__":
    # 输入输出文件路径
    input_file = r"d:/VsCodeProj/RL/SFT/film/test.json"
    output_file = "d:/VsCodeProj/RL/film/sft_test.json"
    
    
    convert_dialogue(input_file, output_file)
    
    print("输出样本预览（前5条）：")
    with open(output_file, 'r', encoding='utf-8') as f:
        for idx, line in enumerate(f):
            if idx >= 5:
                break
            data = json.loads(line)
            print(f"\n【第{idx+1}条】")
            for conv in data["conversations"]:
                print(conv)
```
